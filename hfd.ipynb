```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth =.37
 
# set height of bar
df =  [35,20,42,82,90,37,25,14,19]
df2 = [30,22,26,77,66,25,19,10,15]

 
# Set position of bar on X axis
r1 = np.arange(len(df))
r2 = [x + barWidth for x in r1]

 
# Make the plot
plt.bar(r1, df, color='#1525d4', width=barWidth, edgecolor='white', label='whit')
plt.bar(r2, df2, color='#05962b', width=barWidth, edgecolor='white', label='whitout')

 
# Add xticks on the middle of the group bars
#plt.xlabel('group', fontweight='bold')
plt.ylabel('Percentage', fontweight='bold')
plt.title('Medicine use report of ace-at2')
plt.xticks([r + barWidth for r in range(len(df))], ['any-angina','atrial-fibrillation', 'heart-failure',  'hyperlidemai','hypertension','NIDDM','peripheral Artery disease','previous Stroke','chron-lung-disease'], rotation=90)


# Create legend & Show graphic
plt.legend()
plt.grid(True)
plt.show()

```
